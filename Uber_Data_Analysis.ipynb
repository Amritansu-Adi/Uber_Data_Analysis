```python
def predict_price(name, source, surge_multiplier, icon):
    # Create a zero row with correct column names
    x = pd.DataFrame(np.zeros((1, len(new_uber.columns))), columns=new_uber.columns)

    # Fill the known values
    x.iloc[0, 0] = source
    x.iloc[0, 1] = surge_multiplier
    x.iloc[0, 2] = icon

    # One-hot encode the 'name' column if it exists
    loc_index = np.where(new_uber.columns == name)[0]
    if loc_index.size > 0:
        x.iloc[0, loc_index[0]] = 1

    return random.predict(x)[0]
```
